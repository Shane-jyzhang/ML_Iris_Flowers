```python
Y_iris_df_task_num=iris_df[['target']][50:150]
Y_iris_df_task_num   
```

```python
from sklearn.preprocessing import StandardScaler
x = StandardScaler().fit_transform(X_iris_df_task_num)
df_norm_indata=pd.DataFrame(data=x, columns=iris['feature_names']) 
df_norm_indata['target']=iris['target'][50:150]
df_norm_indata
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principalComponents = pca.fit_transform(x)
principalDf = pd.DataFrame(data = principalComponents
             , columns = ['principal component 1', 'principal component 2'])
principalDf
```

```python
finalDf = pd.concat([principalDf, df_norm_indata[['target']]], axis = 1)
finalDf
```

```python
from matplotlib.colors import ListedColormap
colors = ListedColormap(['r','b','g'])

ax = plt.subplots(figsize=(5, 5))

scatter=plt.scatter(finalDf.iloc[:,0],finalDf.iloc[:,1],c=finalDf['target'],cmap=colors)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
classes=iris['target_names'][1:3].tolist()
plt.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.show()
```

```python
pca_line = PCA().fit(X_iris_df_task_num)
plt.plot([1,2,3,4], np.cumsum(pca_line.explained_variance_ratio_))
plt.xticks([1,2,3,4])
plt.xlabel("number of components after dimension reduction")
plt.ylabel("cumulative explained variance")
plt.show()
```

```python
pca.explained_variance_ratio_.sum()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X_iris_df_task_num, Y_iris_df_task_num, 
                                                    test_size=0.20, random_state=29)
```

```python
#x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(max_depth = 3, random_state = 29)
model.fit(x_train, y_train)
model.score(x_train, y_train)
```

```python
model.score(x_test, y_test)
```

```python
from sklearn import tree
ax = plt.subplots(figsize = (10,10))
tree.plot_tree(model)
plt.show()
```
