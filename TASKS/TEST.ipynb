```python
import pandas as pd     # dataframes
import numpy as np      # numbers and arrays 
import matplotlib.pyplot as plt  # plotting
from sklearn import datasets  
```

```python
iris = datasets.load_iris()
iris
```

```python
iris_df=pd.DataFrame(data=iris['data'], columns=iris['feature_names'])  # iris['data'] is a 2D array 
iris_df
```

```python
iris_df['target']=iris['target']  # add a column to the dataframe called 'target' and insert the 1D array iris['target']
iris_df
```

```python
iris_df_num = iris_df.copy
iris_df_num
```

```python
iris_df = iris_df.drop(index =[0,49], axis=0)
iris_df
```

```python
iris_df['target']=iris_df_num['target'].replace( { 0:iris['target_names'][0] , 1:iris['target_names'][1], 2:iris['target_names'][2]  } )
iris_df
```

```python
iris_df['target']=iris_df['target'].replace( { 0:iris['target_names'][0] , 1:iris['target_names'][1], 2:iris['target_names'][2]  } )
iris_df
```

```python
X_iris_df=iris_df.copy()
X_iris_df=X_iris_df.drop(['target'],axis=1)
X_iris_df
```

```python
plt.scatter(X_iris_df.iloc[:,0],X_iris_df.iloc[:,3])  
plt.show()
```

```python
from matplotlib.colors import ListedColormap
colors = ListedColormap(['r','b','g'])

first_column=0
second_column=3

scatter=plt.scatter(iris_df.iloc[:,first_column],iris_df.iloc[:,second_column],c=iris_df['target'],cmap=colors)
plt.xlabel(iris['feature_names'][first_column])
plt.ylabel(iris['feature_names'][second_column])
classes=iris['target_names'].tolist()
plt.legend(handles=scatter.legend_elements()[0], labels=classes)

iris_df.plot.scatter(x='Sepal length',y='Petal width',c='colors',colormap='viridis')
plt.show()
plt.show()
```
