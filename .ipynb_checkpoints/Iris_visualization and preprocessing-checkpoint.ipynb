```python
import pandas as pd     # dataframes
import numpy as np      # numbers and arrays 
import matplotlib.pyplot as plt  # plotting
from sklearn import datasets                            
```

### Load Iris Data

```python
iris = datasets.load_iris()
iris
```

### Make an Iris DataFrame

```python
iris_df=pd.DataFrame(data=iris['data'], columns=iris['feature_names'])  # iris['data'] is a 2D array 
iris_df
```

```python
iris_df['target']=iris['target']  # add a column to the dataframe called 'target' and insert the 1D array iris['target']
iris_df
```

```python
iris_df_num=iris_df.copy()   # makes an isolated copy
iris_df_num
```

```python
iris['target_names']
```

```python
# replacing target numbers to target names in iris_df
iris_df['target']=iris_df['target'].replace( { 0:iris['target_names'][0] , 1:iris['target_names'][1], 2:iris['target_names'][2]  } )
iris_df
```

```python
iris_df_num # double check
```

```python
X_iris_df_num=iris_df_num.copy()
X_iris_df_num
```

```python
X_iris_df_num=X_iris_df_num.drop(['target'],axis=1)  # drops the target column, only feature columns left
```

```python
X_iris_df_num
```

```python
X_iris_df_num.plot()  # plots each column as a function of index (row number)
plt.show()
```

```python
plt.scatter(X_iris_df_num.iloc[:,1],X_iris_df_num.iloc[:,2])  # X_iris_df_num.iloc[:,1] slice out the second column 
plt.show()
```

```python
from matplotlib.colors import ListedColormap
colors = ListedColormap(['r','b','g'])

first_column=2
second_column=3

scatter=plt.scatter(iris_df_num.iloc[:,first_column],iris_df_num.iloc[:,second_column],c=iris_df_num['target'],cmap=colors)
plt.xlabel(iris['feature_names'][first_column])
plt.ylabel(iris['feature_names'][second_column])
classes=iris['target_names'].tolist()
plt.legend(handles=scatter.legend_elements()[0], labels=classes)
plt.show()
```
